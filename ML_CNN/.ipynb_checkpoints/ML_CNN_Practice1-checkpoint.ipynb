```python
"""

无法解读时候的学习是空中楼阁   白费力气


收获大小

"""
```

```python
import numpy as np
```

```python
def softmax(x):
    x = np.array(x)
    x_exp = np.exp(x)
    x_exp_sum = x_exp.sum()
    y = x_exp/x_exp_sum
    return y
```

## Original

```python
# make the plots
import matplotlib.pyplot as plt

X = [1,2,3,-4,9,8,10,-1,16,-10]
Y = X

plt.bar(range(len(X)), Y)
plt.title('')
plt.ylabel('')
plt.xlabel('')
plt.xticks(range(len(X)))

plt.show()
```

# Softmax

```python
# make the plots
import matplotlib.pyplot as plt

X = [1,2,3,-4,9,8,10,-1,16,-10]
Y = softmax(X)

plt.bar(range(len(X)), Y)
plt.title('')
plt.ylabel('')
plt.xlabel('')
plt.xticks(range(len(X)))

plt.show()
```
