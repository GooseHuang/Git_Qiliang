```python
class Solution:
    def numJewelsInStones(self, jewels: str, stones: str) -> int:
        hash_map = {}
        for x in stones:
            if x in hash_map:
                hash_map[x] += 1
            else:
                hash_map[x] = 1
        
        count = 0
        for x in jewels:
            if x in hash_map:
                count+=hash_map[x]
                
        return count
```
