````python
class Solution:
    def sort_by_length(self, lst: List[str]) -> None:
        """
        Sorts a list of strings by the length of each string
        """        
        lst.sort(key=lambda x: len(x)) # Note we can also do lst.sort(key=len)
````

#### Java

```java
import java.util.Arrays;

public class Solution {
    public void sortByLength(String[] arr) {
        // Sorts a list of string by length of each string
        Arrays.sort(array, new StringCompare());
    }
}

public class StringCompare implements Comparator<String> {
    public int compare(String s1, String s2) {
        if (s1.length() > s2.length()) {
            return 1;
        } else if (s1.length() < s2.length()) {
            return -1;
        }
        return 0;
    }
}
```
