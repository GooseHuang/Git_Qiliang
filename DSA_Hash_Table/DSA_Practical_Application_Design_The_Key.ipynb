```python
class Solution:
    
    def get_index(self, s):
        s = ''.join(sorted(list(s)))
        return s

    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        hash_map = dict()
        for x in strs:
            hash_index = self.get_index(x)
            if hash_index in hash_map:
                hash_map[hash_index].append(x)
            else:
                hash_map[hash_index] = [x,]
        
        output = []
        for k,v in hash_map.items():
            output.append(v)
        return output
```

```python
class Solution:
    
    def shilft(self, x):
        if x == 97:
            return 122
        else:
            return x-1
    
    
    def shilft2parent(self, s):
        s = [ord(x) for x in s]
        while s[0] != 97:
            s = [self.shilft(x) for x in s]
        s = [chr(x) for x in s]
        return ''.join(s)
    
    
    def groupStrings(self, strings: List[str]) -> List[List[str]]:
        hash_map = defaultdict()
        for x in strings:
            hash_index = self.shilft2parent(x)
            if not hash_index in hash_map:
                hash_map[hash_index] = [x,]
            else:
                hash_map[hash_index].append(x)
        output = []
        for k,v in hash_map.items():
            output.append(v)
        return output
```
