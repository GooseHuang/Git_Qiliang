```python
class Solution:
    
    def get_index(self, s):
        s = ''.join(sorted(list(s)))
        return s

    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        hash_map = dict()
        for x in strs:
            hash_index = self.get_index(x)
            if hash_index in hash_map:
                hash_map[hash_index].append(x)
            else:
                hash_map[hash_index] = [x,]
        
        output = []
        for k,v in hash_map.items():
            output.append(v)
        return output
```

```python
class Solution:
    
    def shilft(self, x):
        if x == 97:
            return 122
        else:
            return x-1
    
    
    def shilft2parent(self, s):
        s = [ord(x) for x in s]
        while s[0] != 97:
            s = [self.shilft(x) for x in s]
        s = [chr(x) for x in s]
        return ''.join(s)
    
    
    def groupStrings(self, strings: List[str]) -> List[List[str]]:
        hash_map = defaultdict()
        for x in strings:
            hash_index = self.shilft2parent(x)
            if not hash_index in hash_map:
                hash_map[hash_index] = [x,]
            else:
                hash_map[hash_index].append(x)
        output = []
        for k,v in hash_map.items():
            output.append(v)
        return output
```

```python
class Solution:
    def isValidSudoku(self, board: List[List[str]]) -> bool:
        for i in range(9):
            tmp_set = set()
            for j in range(9):
                if board[i][j] in tmp_set:
                    # print('tata1')
                    return False
                else:
                    if board[i][j] != '.':
                        tmp_set.add(board[i][j])
        
        for j in range(9):
            tmp_set = set()
            for i in range(9):
                if board[i][j] in tmp_set:
                    # print('tata2')
                    return False
                else:
                    if board[i][j] != '.':
                        tmp_set.add(board[i][j])
                        
        for i in range(1,8,3):
            for j in range(1,8,3):
                tmp_set = set()
                check_list = [board[i][j],
                             board[i-1][j],
                             board[i+1][j],
                             board[i][j-1],
                             board[i][j+1],
                             board[i-1][j-1],
                             board[i-1][j+1],
                             board[i+1][j-1],
                             board[i+1][j+1]]
                
                for x in check_list:
                    if x in tmp_set:
                        # print('tata3')
                        # print(x)
                        # print(tmp_set)
                        return False
                    else:
                        if x != '.':
                            tmp_set.add(x)
        return True
        
```
