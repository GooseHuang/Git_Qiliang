##  Reverse String

https://leetcode.com/explore/learn/card/array-and-string/205/array-two-pointer-technique/1183/

```cpp
class Solution {
public:

    void helper(vector<char>& s, int start, int end){
        if(start>=end){
            return;
        }
            
        char tmp = s[start];
        s[start] = s[end];
        s[end] = tmp;
        helper(s, start+1,end-1);
    }
    
    void reverseString(vector<char>& s) {
        helper(s,0,s.size()-1);
    }       
};

```

## Array Partition I
https://leetcode.com/explore/learn/card/array-and-string/205/array-two-pointer-technique/1154/

##### Accepetd with quickSort
```cpp
class Solution {
public:

    void quickSortHelper(vector<int>& nums, int start, int end){
        if(start>=end){
            return;
        }
        
//         cout<<" start: "<<start<<"\n";
//         cout<<" end: "<<end<<"\n";
        
        int pivot = nums[start];
        int i=start;
        int j=end;        
        int cur=i; // point to blank position
        while(i<j){
            if(cur==i){
                if(nums[j]<pivot){
                    nums[i]=nums[j];
                    cur=j;
                    i++;
                }else{
                    j--;
                }
            }else{
                // when cur==j
                if(nums[i]>pivot){
                    nums[j]=nums[i];
                    cur=i;
                    j--;
                }else{
                    i++;
                }
                
            }  
        }
        nums[i] = pivot;
        quickSortHelper(nums,start,i-1);
        quickSortHelper(nums,i+1,end);
        
    }
        
        
   void quickSort(vector<int>& nums){
        quickSortHelper(nums, 0, nums.size()-1);
    }
    
    
    
    int arrayPairSum(vector<int>& nums) {
        quickSortHelper(nums,0,nums.size()-1);
        // for(auto x: nums){
        //     cout<<" x: "<<x<<"\n";
        // }
        
        int s=0;
        for(int i=0;i<nums.size();i+=2){
            s += min(nums[i],nums[i+1]);
        }
        
        return s;
    }
};
```

## Two Sum II - Input array is sorted

https://leetcode.com/explore/learn/card/array-and-string/205/array-two-pointer-technique/1153/

```cpp
class Solution {
public:
    
    
vector<int> twoSum(vector<int>& numbers, int target) {
        
    vector<int> v(2);
    int pivot = numbers.size()-1;
    int c;
    map<int,int> d1;
    
    // cout<<pivot<< " Pivot\n";
        
    for(int i; i<=pivot; i++){
        if(d1.find(numbers[i]) != d1.end()){
            v[1] = i+1;
            v[0] = d1[numbers[i]]+1;
            return v;
        }else{
            c= target-numbers[i];
            d1[c] = i;
            }
    }
    

    return v;
  }

};
```

```cpp
int removeElement(vector<int>& nums, int val) {
    int k = 0;
    for (int i = 0; i < nums.size(); ++i) {
        if (nums[i] != val) {
            nums[k] = nums[i];
            ++k;
        }
    }
    return k;
}
```

## Max Consecutive Ones

https://leetcode.com/explore/learn/card/array-and-string/205/array-two-pointer-technique/1301/

```cpp
class Solution {
public:
    int findMaxConsecutiveOnes(vector<int>& nums) {
        int max=0;
        int accumulate=0;
        for (int i = 0; i < nums.size(); i++){
            if(nums[i]==1){
                accumulate+=1;
                
                if(max < accumulate){
                    max = accumulate;
                    }
                }
            else
                {
                accumulate = 0;
                }
        }
        
        return max;
    }
};
```

##### Accepted devide and conquer
```cpp
class Solution {
public:
    
    int helper(int target, vector<int>& nums, int start, int end){
        if(start==end){
            if(nums[start]>=target){
                return 1;
            }else{
                return INT_MAX;
            }
        }
        
        int mid = start + (end-start)/2;
        
        // cout<<" start: "<<start<<"\n";
        // cout<<" end: "<<end<<"\n";
        // cout<<" mid: "<<mid<<"\n";
        // cout<<"\n";
        
        int left = helper(target, nums, start, mid);
        int right = helper(target, nums, mid+1, end);
        int cross=2;
        int s=nums[mid] + nums[mid+1];
        int i=mid-1;
        int j=mid+2;
        while(s<target){
            if(i<0 and j>end){
                cross=INT_MAX;
                break;
            }else if(i>=0 and j<=end){
                if(nums[i]<nums[j]){
                    s+=nums[j];
                    j++;
                }else{
                    s+=nums[i];
                    i--;
                }
            }else if(i>=0){
                    s+=nums[i];
                    i--;
            }else{
                    s+=nums[j];
                    j++;
            }
            cross++;
        }
        
        int n = min(left,right);
        n = min(n,cross);

        return n;
        
    }
    
    

    int minSubArrayLen(int target, vector<int>& nums) {
        // Devide and conquer
        int n = helper(target, nums, 0, nums.size()-1);
        if(n==INT_MAX){
            n=0;
        }
        return n;
        
    }
};
```
