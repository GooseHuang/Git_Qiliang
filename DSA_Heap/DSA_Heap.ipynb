### Binary Tree Level Order Traversal

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:

    def BFS(self, cur, v, levels):
        if not cur:
            return
        
        if len(v)==levels:
            v.append([]);  
        
        v[levels].append(cur.val);
        
        if cur.left:
            self.BFS(cur.left, v, levels+1)
        
        if cur.right:
            self.BFS(cur.right, v, levels+1)
        
    
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        v = list()
        self.BFS(root, v, 0);
        return v;
```
