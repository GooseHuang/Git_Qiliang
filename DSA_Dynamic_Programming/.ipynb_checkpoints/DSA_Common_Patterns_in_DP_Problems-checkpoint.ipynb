```python
# Top-Bottom I
class Solution:
    def minDifficulty(self, jobDifficulty: List[int], d: int) -> int:
        n = len(jobDifficulty)
        # If we cannot schedule at least one job per day, 
        # it is impossible to create a schedule
        if n < d:
            return -1
        
        hardest_job_remaining = [0] * n
        hardest_job = 0
        for i in range(n - 1, -1, -1):
            hardest_job = max(hardest_job, jobDifficulty[i])
            hardest_job_remaining[i] = hardest_job
        
        @lru_cache(None)
        def dp(i, day):
            # Base case, it's the last day so we need to finish all the jobs
            if day == d:
                return hardest_job_remaining[i]
            
            best = float("inf")
            hardest = 0
            # Iterate through the options and choose the best
            for j in range(i, n - (d - day)): # Leave at least 1 job per remaining day
                hardest = max(hardest, jobDifficulty[j])
                best = min(best, hardest + dp(j + 1, day + 1)) # Recurrence relation

            return best
        
        return dp(0, 1)
```

```python
# Accepted
class Solution:
    def coinChange(self, coins: List[int], amount: int) -> int:
        
        @lru_cache(None)
        def dp(amount):
            
            if amount < 0:
                return float("inf")
            
            if amount == 0:
                return 0
            
            if amount in coins:
                return 1
            
            
            best = float("inf")
            for i in coins:
                best = min(best, 1+ dp(amount-i))
            return best
        
        result = dp(amount)
        if result == float("inf"):
            result = -1
        
        return result
```

```python
class Solution:
    def wordBreak(self, s: str, wordDict: List[str]) -> bool:
        
        @lru_cache(None)
        def dp(s):
            
            if not s:
                return True
            
            if s in wordDict:
                return True
            
            
            target_list = []
            for i in range(len(s)):
                if s[:i+1] in wordDict:
                    target_list.append(dp(s[i+1:]))
                    
            return any(target_list)
        
        return dp(s)
```

```python
# Failed v1
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        @lru_cache(None)
        def dp(min_num, start):
            while start < len(nums) and min_num >= nums[start]:
                start += 1
            if start >= len(nums):
                return 0
            elif start == len(nums)-1:
                return 1
            else:
                cur = nums[start]
                next_cur = nums[start+1]
                max_num = max(1+dp(cur,start+1), dp(next_cur, start+2))
                return max_num
        return max(1+dp(nums[0], 1), dp(nums[1], 2))
```

```python
# Time Exceeded v2
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        
        @lru_cache(None)
        def dp(min_num, start):
            while start < len(nums) and min_num >= nums[start]:
                start += 1
            if start >= len(nums):
                return 0
            elif start == len(nums)-1:
                if nums[start] > min_num:
                    return 1
                else:
                    return 0
            else:
                cur = nums[start]
                max_num = max(1+dp(cur,start+1), dp(min_num,start+1))
                return max_num
            
        return dp(-float('inf'), 0)
```

```python
# Accepted BP
class Solution:
    def lengthOfLIS(self, nums: List[int]) -> int:
        dp = [1] * len(nums)
        for i in range(1, len(nums)):
            for j in range(i):
                if nums[i] > nums[j]:
                    dp[i] = max(dp[i], dp[j] + 1)

        return max(dp)
```

```python
# Accepted V1
class Solution:        
    def maxProfit(self, k: int, prices: List[int]) -> int:
        @lru_cache(None)
        def dp(i, transactions_remaining, holding):
            # Base case
            if transactions_remaining == 0 or i == len(prices): 
                return 0
            
            do_nothing = dp(i + 1, transactions_remaining, holding)
            #do_something = 0

            if holding:
                # Sell stock
                do_something = prices[i] + dp(i + 1, transactions_remaining - 1, 0)
            else:
                # Buy stock
                do_something = -prices[i] + dp(i + 1, transactions_remaining, 1)

            # Recurrence relation
            return max(do_nothing, do_something)

        return dp(0, k, 0)
```

```python
# Method 2
class Solution:
    def maxProfit(self, k: int, prices: List[int]) -> int:
        n = len(prices)

        # solve special cases
        if not prices or k==0:
            return 0

        if 2*k > n:
            res = 0
            for i, j in zip(prices[1:], prices[:-1]):
                res += max(0, i - j)
            return res

        # dp[i][used_k][ishold] = balance
        # ishold: 0 nothold, 1 hold
        dp = [[[-math.inf]*2 for _ in range(k+1)] for _ in range(n)]

        # set starting value
        dp[0][0][0] = 0
        dp[0][1][1] = -prices[0]

        # fill the array
        for i in range(1, n):
            for j in range(k+1):
                # transition equation
                dp[i][j][0] = max(dp[i-1][j][0], dp[i-1][j][1]+prices[i])
                # you can't hold stock without any transaction
                if j > 0:
                    dp[i][j][1] = max(dp[i-1][j][1], dp[i-1][j-1][0]-prices[i])

        res = max(dp[n-1][j][0] for j in range(k+1))
        return res
```

```python
# Memory Exceeded
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        @lru_cache(None)
        def dp(i, last_sell_day, holding):
            # Base case
            if i == len(prices): 
                return 0
            
            do_nothing = dp(i + 1, last_sell_day, holding)
            do_something = 0

            if holding:
                # Sell stock
                do_something = prices[i] + dp(i + 1, i, 0)
            else:
                if last_sell_day+1!=i:
                    # Buy stock
                    do_something = -prices[i] + dp(i + 1, last_sell_day, 1)

            # Recurrence relation
            return max(do_nothing, do_something)

        return dp(0, -2 ,0)
```
