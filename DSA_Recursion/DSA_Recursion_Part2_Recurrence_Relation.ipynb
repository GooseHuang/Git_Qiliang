/**
 * Definition for singly-linked list.
 * struct ListNode {
 *     int val;
 *     ListNode *next;
 *     ListNode() : val(0), next(nullptr) {}
 *     ListNode(int x) : val(x), next(nullptr) {}
 *     ListNode(int x, ListNode *next) : val(x), next(next) {}
 * };
 */
class Solution {
public:
    
    ListNode* helper(ListNode* pre, ListNode* cur){
        if(!cur->next){
            // cout<<cur->val<<" cur->val\n";
            // cout<<pre->val<<" pre->val\n";
            // cout<<pre->next->val<<" pre->next->val\n";
            cur->next = pre;
            // cout<<" tata2\n";
            return cur;
        }else{
            // cout<<cur->val<<" cur->val\n";
            // cout<<cur->next->val<<" cur->next->val\n";
            ListNode* head = helper(cur, cur->next);
            cur->next = pre;
            // cout<<" tata\n";
            return head;
            }
        }
    
    
    ListNode* reverseList(ListNode* head) {
        if(!head or !head->next){
            return head;
        }
        
        ListNode* tmp = head->next;
        head->next = NULL;
        head = helper(head, tmp);
    return head;
    
    }    
};

public ListNode reverseList(ListNode head) {
    ListNode prev = null;
    ListNode curr = head;
    while (curr != null) {
        ListNode nextTemp = curr.next;
        curr.next = prev;
        prev = curr;
        curr = nextTemp;
    }
    return prev;
}
