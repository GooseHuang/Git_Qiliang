````python
class Solution:
    def bubble_sort(self, lst: List[int]) -> None:
        """
        Mutates lst so that it is sorted via swapping adjacent elements until
        the entire lst is sorted.
        """
        has_swapped = True
        # if no swap occurred, lst is sorted
        while has_swapped:
            has_swapped = False
            for i in range(len(lst) - 1):
                if lst[i] > lst[i + 1]:
                    # Swap adjacent elements
                    lst[i], lst[i + 1] = lst[i + 1], lst[i]
                    has_swapped = True          
````

````python
# Insertion Sort
class Solution:
    def insertion_sort(self, lst: List[int]) -> None:
        """
        Mutates elements in lst by inserting out of place elements into appropriate
        index repeatedly until lst is sorted
        """
        for i in range(1, len(lst)):
            current_index = i

            while current_index > 0 and lst[current_index - 1] > lst[current_index]:
                # Swap elements that are out of order
                lst[current_index], lst[current_index - 1] = lst[current_index - 1], lst[current_index]
                current_index -= 1
````

````python
# Great Ways     No difference whether start from left to right   Easily

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def insertionSortList(self, head: ListNode) -> ListNode:
        dummy = ListNode()
        curr = head

        while curr:
            # At each iteration, we insert an element into the resulting list.
            prev = dummy

            # find the position to insert the current node
            while prev.next and prev.next.val <= curr.val:
                prev = prev.next

            next = curr.next
            # insert the current node to the new list
            curr.next = prev.next
            prev.next = curr

            # moving on to the next iteration
            curr = next

        return dummy.next
````

````python
class Solution:
    def heap_sort(self, lst: List[int]) -> None:
        """
        Mutates elements in lst by utilizing the heap data structure
        """
        def max_heapify(heap_size, index):
            left, right = 2 * index + 1, 2 * index + 2
            
            largest = index
            
            if left < heap_size and lst[left] > lst[largest]:
                largest = left
                
            if right < heap_size and lst[right] > lst[largest]:
                largest = right
                
            if largest != index:
                lst[index], lst[largest] = lst[largest], lst[index]
                max_heapify(heap_size, largest)

        # heapify original lst
        for i in range(len(lst) // 2 - 1, -1, -1):
            max_heapify(len(lst), i)

        # use heap to sort elements
        for i in range(len(lst) - 1, 0, -1):
            # swap last element with first element
            lst[i], lst[0] = lst[0], lst[i]
            # note that we reduce the heap size by 1 every iteration
            max_heapify(i, 0)
````

````python
# Selection Sort
class Solution:
    def selection_sort(self, lst: List[int]) -> None:
        """
        Mutates lst so that it is sorted via selecting the minimum element and
        swapping it with the corresponding index
        """
        for i in range(len(lst)):
            min_index = i
            for j in range(i + 1, len(lst)):
                # Update minimum index
                if lst[j] < lst[min_index]:
                    min_index = j

            # Swap current index with minimum element in rest of list
            lst[min_index], lst[i] = lst[i], lst[min_index]
````

````python
# Implementation of a Heap
    
class Heap:
    def __init__(self,size=8):
        self.heap = [None]*size
        self.size = 0
        
    def add(self,val):
        self.size+=1
        if self.size>=len(self.heap):
            self.up_size()
        
        self.heap[self.size-1] = val
        self.up_heapify()

    def pop(self):
        if self.size==0:
            return None
        
        
        i = self.size-1
        res = self.heap[0]
        self.heap[0] = self.heap[i]
        self.heap[i] = None
        self.size-=1
        if self.size<(len(self.heap)//2):
            self.down_size()
        self.down_heapify()
        return res
        
        
    def get(self):
        return self.heap[0]
    
    def up_heapify(self):
        child_i = self.size-1
        parent_i = (child_i-1)//2
        while(parent_i>=0):
            
            parent = self.heap[parent_i]
            child = self.heap[child_i]
            
            if parent>child:
                self.heap[parent_i] = child
                self.heap[child_i] = parent
                child_i = parent_i
                parent_i = (child_i-1)//2
            else:
                break
                
                
    def down_heapify(self):
        parent_i = 0
        while(parent_i<self.size-1):
            left_i = parent_i*2+1
            right_i = parent_i*2+2
            
            parent = self.heap[parent_i]
            
            if right_i<self.size:    
                left = self.heap[left_i]
                right = self.heap[right_i]

                if (parent<=left) and (parent<=right):   
                    break
                if left>right:
                    self.heap[parent_i] = right
                    self.heap[right_i] = parent
                    parent_i = right_i
                else:
                    self.heap[parent_i] = left
                    self.heap[left_i] = parent
                    parent_i = left_i
            elif left_i<self.size:
                left = self.heap[left_i]
                if (parent<=left):
                    break
                else:
                    self.heap[parent_i] = left
                    self.heap[left_i] = parent
                    parent_i = left_i
            else:
                break    

    def up_size(self):
        new_heap = [None] * (self.size*2)
        for i,x in enumerate(self.heap):
            new_heap[i] = x
        self.heap = new_heap
        
    def down_size(self):
        new_heap = [None] * (len(self.heap)//2)
        for i in range(self.size):
            new_heap[i] = self.heap[i]
        self.heap = new_heap
        
        
        
heap = Heap()
heap.add(1)
heap.add(2)
heap.add(3)
heap.add(4)
heap.add(5)
heap.add(6)
heap.add(7)
heap.add(0)

print(heap.pop())
heap.add(-1)
print(heap.pop())
print(heap.heap)
````

````python
# Merge Sort
def merge_sort(target_list):
    
    def split(target_list):
        pivot_i = len(target_list)//2
        listA = target_list[:pivot_i]
        listB = target_list[pivot_i:]
        return listA,listB
    
    def merge(listA,listB):
        res_list = []
        i = 0
        j = 0
        while(i<len(listA) and j<len(listB)):
            a = listA[i]
            b = listB[j]
            if a<b:
                res_list.append(a)
                i+=1
            else:
                res_list.append(b)
                j+=1
                
        while(i<len(listA)):
            a = listA[i]
            res_list.append(a)
            i+=1
            
        while(j<len(listB)):
            b = listB[j]
            res_list.append(b)
            j+=1

        return res_list
    
    def sort(target_list):
        if len(target_list)<=1:
            return target_list
        elif len(target_list)==2:
            if target_list[0]>target_list[1]:
                return [target_list[1],target_list[0]]
            else:
                return target_list
        else:
            listA,listB = split(target_list)
            listA = sort(listA)
            listB = sort(listB)
            res_list = merge(listA,listB)
            return res_list
    
    
    return sort(target_list)
    
import random
target_list = random.sample(range(100),k=10)
# target_list = [41, 39, 80, 97, 44, 52, 88, 58, 59, 57]
print(target_list)
target_list = merge_sort(target_list)
target_list    
````
