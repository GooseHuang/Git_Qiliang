```python
class Solution:
    def numJewelsInStones(self, jewels: str, stones: str) -> int:
        hash_map = {}
        for x in stones:
            if x in hash_map:
                hash_map[x] += 1
            else:
                hash_map[x] = 1
        
        count = 0
        for x in jewels:
            if x in hash_map:
                count+=hash_map[x]
                
        return count
```

```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        
        if len(s)==1:
            return 1
        
        max_length = 0
        for i in range(len(s)):
            hash_set = set()
            j = i
            while j < len(s):
                # print(s[j])
                if not s[j] in hash_set:
                    hash_set.add(s[j])
                    j += 1
                    # print(max_length)
                else:
                    if len(hash_set) > max_length:
                        max_length = len(hash_set) 
                        # print(hash_set)
                    break
                
            if j==len(s):
                if len(hash_set) > max_length:
                    max_length = len(hash_set) 
                break
                
                
            
        return max_length
```
