# DSA_Array_and_String_Part3_Introduction_to_String

A string is actually an array of unicode characters. You can perform almost all the operations we used in an array. You can try it out by yourself.

However, there are some differences. In this article, we will go through some of them which you should be aware of when dealing with a string. These features might vary a lot from one language to another. 

```cpp
#include <iostream>

int main() {
    string s1 = "Hello World";
    cout << "s1 is \"Hello World\"" << endl;
    string s2 = s1;
    cout << "s2 is initialized by s1" << endl;
    string s3(s1);
    cout << "s3 is initialized by s1" << endl;
    // compare by '=='
    cout << "Compared by '==':" << endl;
    cout << "s1 and \"Hello World\": " << (s1 == "Hello World") << endl;
    cout << "s1 and s2: " << (s1 == s2) << endl;
    cout << "s1 and s3: " << (s1 == s3) << endl;
    // compare by 'compare'
    cout << "Compared by 'compare':" << endl;
    cout << "s1 and \"Hello World\": " << !s1.compare("Hello World") << endl;
    cout << "s1 and s2: " << !s1.compare(s2) << endl;
    cout << "s1 and s3: " << !s1.compare(s3) << endl;
}

```

##  Add Binary
https://leetcode.com/explore/learn/card/array-and-string/203/introduction-to-string/1160/

##### Accepted v1
```cpp
class Solution {
public:
    
    int translate(char s){
        if(s=='1'){
            return 1;
        }else{
            return 0;
        }
    }
    
    
    string addBinary(string a, string b) {
        int len_a = a.size();
        int len_b = b.size();
        int size = max(len_a,len_b);
        int i = 0;
        int step = 0;
        int cur = 0;
        vector<int> output; 
        while(i<size){
            if(i>=len_a){
              cur = translate(b[len_b-1-i]) + step;
            }else if(i>=len_b){
              cur = translate(a[len_a-1-i]) + step;
            }else{
              cur = translate(a[len_a-1-i]) + translate(b[len_b-1-i]) + step;
            }
            
            if(cur==2){
                cur=0;
                step=1;
            }else if(cur==3){
                cur=1;
                step=1;
            }else{
                step=0;
            }
            
            output.push_back(cur);
            i++;
        }
        
        if(step==1){
            output.push_back(1);
        }
        string output_string;
        for(int i=output.size()-1;i>=0;i--){
            int cur=output[i];
            // cout<<"cur: "<<cur<<"\n";
            if(cur==1){
                output_string+='1';
            }else{
                output_string+='0';
            }
        }
        
        return output_string;
        
        
        
        
        
    }
};

```

### Solution
https://leetcode.com/problems/add-binary/solution/

##### Approach 1: Bit-by-Bit Computation

```java
class Solution {
  public String addBinary(String a, String b) {
    int n = a.length(), m = b.length();
    if (n < m) return addBinary(b, a);
    int L = Math.max(n, m);

    StringBuilder sb = new StringBuilder();
    int carry = 0, j = m - 1;
    for(int i = L - 1; i > -1; --i) {
      if (a.charAt(i) == '1') ++carry;
      if (j > -1 && b.charAt(j--) == '1') ++carry;

      if (carry % 2 == 1) sb.append('1');
      else sb.append('0');

      carry /= 2;
    }
    if (carry == 1) sb.append('1');
    sb.reverse();

    return sb.toString();
  }
}
```

##### Approach 2: Bit Manipulation
```python
class Solution:
    def addBinary(self, a, b) -> str:
        x, y = int(a, 2), int(b, 2)
        while y:
            x, y = x ^ y, (x & y) << 1
        return bin(x)[2:]
```

## Implement strStr()

https://leetcode.com/explore/learn/card/array-and-string/203/introduction-to-string/1161/

##### Accepted v1
```cpp
class Solution {
public:
    int strStr(string haystack, string needle) {
        if(needle==""){
            return 0;
        }

        int i=0;
        int j=i+needle.size()-1;
        
        while(j<haystack.size()){
            string cur = haystack.substr(i,needle.size());
            // cout<<" cur: "<<cur<<"\n";
            if(cur==needle){
                return i;
            }
            i++;
            j=i+needle.size()-1;
        }
        
        return -1;
        
        
    }
};
```

```cpp
class Solution {
public:
    
    int strStr(string haystack, string needle) {
        if(needle==""){
            return 0;
        }

        int i=0;
        int j=i+needle.size()-1;

        while(j<haystack.size()){
            string cur = haystack.substr(i,needle.size());
            // cout<<" cur: "<<cur<<"\n";
            if(cur==needle){
                return i;
            }
            i++;
            j=i+needle.size()-1;
        }

        return -1;

    }

    
    string longestCommonPrefix(vector<string>& strs) {
        string shotest_string = strs[0];
        for(int i=1;i<strs.size();i++){
            if(strs[i].size()<shotest_string.size()){
                shotest_string = strs[i];
            }
        }
        
        int needle_len = 1;
        set<string> needle_set;
        for(char s: shotest_string){
            
            bool validation = true;
            for(string str:strs){
                int needle_position = strStr(str,string(1, s));
                if(needle_position==-1){
                    validation = false;
                    break;
                }
            }
            if(validation){
                needle_set.insert(string(1, s));
            }
        }
        
        while(true){
            set<string>tmp_needle_set;
            for(string needle: needle_set){
                int needle_position = strStr(shotest_string,needle);
                if(needle_position+(needle_len+1)-1 >= shotest_string.size()){
                    continue;
                }
                needle = shotest_string.substr(needle_position,needle_len+1);
                bool validation = true;
                for(string str : strs){
                    int cur = strStr(str, needle);
                    if(cur==-1){
                        validation = false;
                        break;
                    }
                }
                if(validation){
                    tmp_needle_set.insert(needle);
                } 
            }
            
            if(tmp_needle_set.size()==0){
                break;
            }else{
                needle_set = tmp_needle_set;
                needle_len++;
            }  
            
        }
        
        if(needle_set.size()==0){
            return "";
        }
        
        
        
        string output;
        for(string needle:needle_set){
            output = needle;
            break;
        }
        
        return output;
        
        
    }
};

```

##### Approach 1: Horizontal scanning
```java
 public String longestCommonPrefix(String[] strs) {
    if (strs.length == 0) return "";
    String prefix = strs[0];
    for (int i = 1; i < strs.length; i++)
        while (strs[i].indexOf(prefix) != 0) {
            prefix = prefix.substring(0, prefix.length() - 1);
            if (prefix.isEmpty()) return "";
        }        
    return prefix;
}

```

##### Approach 2: Vertical scanning

```java
public String longestCommonPrefix(String[] strs) {
    if (strs == null || strs.length == 0) return "";
    for (int i = 0; i < strs[0].length() ; i++){
        char c = strs[0].charAt(i);
        for (int j = 1; j < strs.length; j ++) {
            if (i == strs[j].length() || strs[j].charAt(i) != c)
                return strs[0].substring(0, i);             
        }
    }
    return strs[0];
}

```

##### Approach 3: Divide and conquer
```java
public String longestCommonPrefix(String[] strs) {
    if (strs == null || strs.length == 0) return "";    
        return longestCommonPrefix(strs, 0 , strs.length - 1);
}

private String longestCommonPrefix(String[] strs, int l, int r) {
    if (l == r) {
        return strs[l];
    }
    else {
        int mid = (l + r)/2;
        String lcpLeft =   longestCommonPrefix(strs, l , mid);
        String lcpRight =  longestCommonPrefix(strs, mid + 1,r);
        return commonPrefix(lcpLeft, lcpRight);
   }
}

String commonPrefix(String left,String right) {
    int min = Math.min(left.length(), right.length());       
    for (int i = 0; i < min; i++) {
        if ( left.charAt(i) != right.charAt(i) )
            return left.substring(0, i);
    }
    return left.substring(0, min);
}

```

##### Approach 4: Binary search
```java
public String longestCommonPrefix(String[] strs) {
    if (strs == null || strs.length == 0)
        return "";
    int minLen = Integer.MAX_VALUE;
    for (String str : strs)
        minLen = Math.min(minLen, str.length());
    int low = 1;
    int high = minLen;
    while (low <= high) {
        int middle = (low + high) / 2;
        if (isCommonPrefix(strs, middle))
            low = middle + 1;
        else
            high = middle - 1;
    }
    return strs[0].substring(0, (low + high) / 2);
}

private boolean isCommonPrefix(String[] strs, int len){
    String str1 = strs[0].substring(0,len);
    for (int i = 1; i < strs.length; i++)
        if (!strs[i].startsWith(str1))
            return false;
    return true;
}
```

```cpp
public String longestCommonPrefix(String q, String[] strs) {
    if (strs == null || strs.length == 0)
         return "";  
    if (strs.length == 1)
         return strs[0];
    Trie trie = new Trie();      
    for (int i = 1; i < strs.length ; i++) {
        trie.insert(strs[i]);
    }
    return trie.searchLongestPrefix(q);
}

class TrieNode {

    // R links to node children
    private TrieNode[] links;

    private final int R = 26;

    private boolean isEnd;

    // number of children non null links
    private int size;    
    public void put(char ch, TrieNode node) {
        links[ch -'a'] = node;
        size++;
    }

    public int getLinks() {
        return size;
    }
    //assume methods containsKey, isEnd, get, put are implemented as it is described
   //in  https://leetcode.com/articles/implement-trie-prefix-tree/)
}

public class Trie {

    private TrieNode root;

    public Trie() {
        root = new TrieNode();
    }

//assume methods insert, search, searchPrefix are implemented as it is described
//in  https://leetcode.com/articles/implement-trie-prefix-tree/)
    private String searchLongestPrefix(String word) {
        TrieNode node = root;
        StringBuilder prefix = new StringBuilder();
        for (int i = 0; i < word.length(); i++) {
            char curLetter = word.charAt(i);
            if (node.containsKey(curLetter) && (node.getLinks() == 1) && (!node.isEnd())) {
                prefix.append(curLetter);
                node = node.get(curLetter);
            }
            else
                return prefix.toString();

         }
         return prefix.toString();
    }
}

```
