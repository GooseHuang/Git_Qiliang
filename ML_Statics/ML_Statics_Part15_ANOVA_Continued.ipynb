```python
import numpy as np


def getSampleStd(array):
    return (((array-array.mean())**2).sum()/(len(array)-1))**0.5


def getCoSampleStd(array_a, array_b):
    return (getSampleStd(array_a)**2/len(array_a) + getSampleStd(array_b)**2/len(array_b))**0.5


def getTScore(array_a,array_b):
    return (array_a.mean()-array_b.mean())/getCoSampleStd(array_a, array_b)

def getRScore(array_a, array_b):
    t_score = getTScore(array_a, array_b)
    Df = len(array_a)-1 + len(array_b)-1 
    return t_score**2/(t_score**2 + Df)

def getArray(df_col):
    return np.array([x for x in df_col if not np.isnan(x)])

def getVariance(array):
    d_array = array - array.mean()
    return (d_array*d_array).sum()

def getPooledVarance(x,y):
    return (getVariance(x)+getVariance(y))/(len(x)-1+len(y)-1)


def getPooledSampleStd(array_a, array_b):
    pooled_variance = getPooledVarance(array_a,array_b)
    return (pooled_variance/len(array_a) + pooled_variance/len(array_b))**0.5

def getPooledTScore(array_a,array_b):
    return (array_a.mean()-array_b.mean())/getPooledSampleStd(array_a, array_b)

```
